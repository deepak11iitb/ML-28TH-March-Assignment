```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})


df.head()
```

```python
## hence the value of flightnumber is incresed by 10 with each row hence its liner so liner interpolation can be used

df["FlightNumber"] = df["FlightNumber"].interpolate(method ='linear', limit_direction ='forward').astype(int)

df
```

```python
# Created new DataFrame df1 with coloums "From" and "To"

df1 = pd.DataFrame()
df1["From"] = df["From_To"].apply(lambda x : x[:x.index('_')].title())
df1["To"] = df["From_To"].apply(lambda x : x[:x.index('_')].title())

df1

```

```python
# Removed "From_To" column from df
df = df.drop("From_To" , axis = 1)
df
```

```python
# joined two dataframe(df1 and df) using concat

df = pd.concat([df1,df],axis = 1)
```

```python
# Created three more delay columns: delay_1,delay_2,delay_3
df["RecentDelays"] = df["RecentDelays"].apply(lambda x : x + [np.NaN,np.NaN,np.NaN])
df["delay_1"] = df["RecentDelays"].apply(lambda x : x[0])
df["delay_2"] = df["RecentDelays"].apply(lambda x : x[1])
df["delay_3"] = df["RecentDelays"].apply(lambda x : x[2])

df
```

```python
#removed the "RecentDelays" column from df
df = df.drop("RecentDelays", axis =1)
df
```
